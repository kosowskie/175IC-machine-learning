```
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split

data = {
    'size': [np.nan, 'L', 'M', 'L', 'S', np.nan, 'L', 'S', 'L', 'M', 'S', 'XL', 'M', 'S', 'L', 'S', 'M', 'L', 'S', 'L'],
    'color': [np.nan, 'green', 'blue', 'green', 'red', 'yellow', np.nan, 'blue', 'green', 'black', 'blue', 'yellow', 'green', 'green', 'black', 'red', 'yellow', 'grey', 'blue', 'gray'],
    'gender': ['male', 'male', 'male', np.nan, np.nan, np.nan, np.nan, 'male', 'male', 'male', 'male', np.nan, 'female', 'female', 'female', 'male', 'female', 'male', 'male', 'male'],
    'price': [np.nan, 89.0, 99.0, 129.0, np.nan, 230.0, np.nan, 180.0, np.nan, 80.0, np.nan, 189.0, 79.0, 179.0, 29.0, 130.0, 180.0, 110.0, 46.0, 89.0],
    'weight': [np.nan, 450, 300, np.nan, np.nan, 700, 630, np.nan, 300, 240, 320, 450, 400, 630, 240, 100, 780, 610, 660, 300],
    'bought': ['yes', 'no', 'yes', 'no', np.nan, 'yes', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'no']
}

df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
```

```
#1
df.dropna(thresh = (len(df.keys())-2), inplace=True)
df
```

```
#2
df['price'].fillna(df['price'].mean(), inplace=True)
df
```

```
#3
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
df
```

```
#4
df['size'].fillna('unisize', inplace=True)
df['color'].fillna('black', inplace=True)
df['gender'].fillna('diverse', inplace=True)
```

```
#5
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
df
```

```
#6
df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
df
```

```
#7
df1=df[['price','cost']]
df2=df['bought_yes']

X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.2)
```
